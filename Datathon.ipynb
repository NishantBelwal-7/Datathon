```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
class bcolors:
    OK = '\033[92m' #GREEN
    WARNING = '\033[93m' #YELLOW
    FAIL = '\033[91m' #RED
    RESET = '\033[0m' #RESET COLOR
```

```python
data = pd.read_csv('Dataset/Internantional dataset.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data['Country_Region'].unique()
```

```python
india = data.query('Country_Region == "India"')
```

```python
india.head()
```

```python
india.info()
```

```python
india.dropna(inplace=True)
```

```python
india.info()
```

```python
india_per = (india[india.columns[2:]] / 
                  india[india.columns[2:]].sum()) * 100
```

```python
india_scaled = pd.DataFrame(india_per,columns = india.columns[2:])
```

```python
states = np.array(india['Province_State'])
```

```python
india_scaled['States'] = states
```

```python
india_scaled = india_scaled.sort_values(by='Confirmed')
```

```python
india_scaled.plot(x="States", y=["Confirmed", "Deaths",'Active','Recovered'], kind="bar", figsize=(20, 8))
```

# Plot which shows the Confirmed, Active and Deaths(in % of all total cases) of all the states in India
### 

```python
top10 = india_scaled.sort_values(by='Confirmed')
top10 = top10[-15:]
top10.plot(x="States", y=["Confirmed", "Deaths",'Active','Recovered'], kind="bar", figsize=(20, 8))
```

### The top 10 states in terms of Confirmed cases
# 

```python
india_scaled = india_scaled.sort_values(by='Case_Fatality_Ratio')
india_scaled.plot(x="States", y=['Incident_Rate','Case_Fatality_Ratio'], kind="bar", figsize=(20, 8))
```

# Plot which show the Incident Rate and Case Fatality Ratio of All the States of India
## 

```python
india_scaled = india_scaled.sort_values(by='Confirmed')
dat = india_scaled[-6:]

labels = list(dat['States'])
labels.append('Others')

print(f"{bcolors.WARNING}\n\t\t!!! Confirmed Cases !!!\n{bcolors.RESET}")
sizes = list(dat['Confirmed'])
sizes.append(100-dat['Confirmed'].sum())
colors = ( "orange", "cyan", "brown", 
          "grey","red",'blue','green') 
fig1, ax1 = plt.subplots()
explode = (0.2, 0.2, 0.2, 0.2, 0.1, 0.1,0.1) 
ax1.pie(sizes, colors = colors, explode=explode, labels=labels,autopct='%1.1f%%',
        shadow=True, startangle=90)

plt.show()
```

### Taking the top 6 states in terms of Confirmed Cases
# 

```python
india_scaled = india_scaled.sort_values(by='Deaths')
dat = india_scaled[-6:]

labels = list(dat['States'])
labels.append('Others')
print(f"{bcolors.FAIL}\n\t\t!!! Deaths !!!\n{bcolors.RESET}")
sizes = list(dat['Deaths'])
sizes.append(100-dat['Deaths'].sum())
colors = ( "orange", "cyan", "brown", 
          "grey","red",'blue','green') 
fig1, ax1 = plt.subplots()
explode = (0.2, 0.2, 0.2, 0.1, 0.1, 0.09,0.05) 
ax1.pie(sizes, colors = colors, explode=explode, labels=labels,autopct='%1.1f%%',
        shadow=True, startangle=90)

plt.show()
```
